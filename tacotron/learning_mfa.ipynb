```python
from praatio import tgio
import sounddevice as sd
import parselmouth
import matplotlib.pyplot as plt
import time
from threading import Thread

# TextGrid 파일과 음성 파일 로드
tg = tgio.openTextgrid("./prosodylab_corpus_directory_aligned/speaker1/LJ001-0003.TextGrid")
sound = parselmouth.Sound("./prosodylab_corpus_directory/speaker1/LJ001-0003.wav")

# TextGrid에서 첫 번째 tier 추출
tiers = tg.tierNameList
data = {}

start_times = []
end_times = []
labels = []

for tier_name in tiers:
    tier = tg.tierDict[tier_name]
    intervals = tier.entryList
    start_times = [interval.start for interval in intervals]
    end_times = [interval.end for interval in intervals]
    labels = [interval.label for interval in intervals]

    break

# 현재 시간을 추적할 변수 초기화
current_time = 0

def play_sound():
    sd.play(sound.values.T, sound.sampling_frequency)
    sd.wait()

def update_labels():
    global current_time
    for start, end, label in zip(start_times, end_times, labels):
        
        while current_time < start:
            time.sleep(0.1)
            current_time += 0.1
        
        if label != '<eps>':
            print(f"Current word: {label}", flush=True)

# 스레드 생성 및 시작
sound_thread = Thread(target=play_sound)
label_thread = Thread(target=update_labels)

sound_thread.start()
label_thread.start()

sound_thread.join()
label_thread.join()
```
